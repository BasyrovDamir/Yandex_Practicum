### Посмотрим, сколько игр выпускалось в разные годы. Важны ли данные за все периоды?

```python
year_of_release_games = pd.pivot_table(data, index = 'year_of_release', values = 'name', aggfunc='count')
year_of_release_games.columns = ['count']
year_of_release_games
```

```python
year_of_release_games.plot(grid = True)

plt.title("График зависимости количества игр от года выпуска")
plt.xlabel("Год выпуска игр")
plt.ylabel("Количество игр")
plt.show()
```

###### Как мы видим по графику, 1980-1993 игры выпускались относительно стабильно с небольшим ростом. С 1994 года появляется резкий скачок, это обусловлено тем, что у большинства населения земли появляются компьютеры и игровые платформы и игры начинают выпускать на cd-формате. По графику видим, что пик наступил 2008-2009 году, обусловлено тем, что к этому времени платформы набрали популярность. После 2009 года начался резкий спад, полагаю это связано с тем, что рынок игр перераспределился в сторону мобильных платформ(ios, androind), а так же браузерные игры.

### Посмотрим, как менялись продажи по платформам. Выберем платформы с наибольшими суммарными продажами и построим распределение по годам.

```python
platforms_game_sales = pd.pivot_table(data, index = 'platform', values = 'total', aggfunc='sum').sort_values(by='total', ascending=False)
platforms_game_sales
```

```python
plt.figure(figsize=(14,5))
sns.barplot(x=platforms_game_sales.index,y=platforms_game_sales['total'], color = 'blue')
plt.title("Продажи по платформам за весь период")
plt.xlabel("Название платформы")
plt.ylabel("Количество продаж")
```

```python
platform_grouped = platforms_game_sales.reset_index()
```

```python
#Составим список всех платформ.
top_platforms_list = platform_grouped['platform'].unique()
top_platforms_list
```

```python
#Построим распределения по годам для каждой платформы. Cоздадим список
for platform in top_platforms_list:
    data[data['platform'] == platform].pivot_table(index = 'year_of_release', values = 'total', aggfunc='sum').plot(kind='bar', figsize=(13,4))
    plt.title(platform)
```

```python
# Создадим свою таблицу по платформам и их обшим продажам. отсортируем их по убыванию и оставим только топ 5.
top_5_platforms = data.pivot_table(index='platform', values='total', aggfunc='sum').sort_values(
    by='total', ascending=False).head(5)
top_5_platforms = top_5_platforms.reset_index().rename_axis(None, axis=1)
```

```python
# Выведем топ 5 продаваемых платформ
top_5_platforms
```

## Проверим гипотезы

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

- Стоит задача проверить две гипотезы:

- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.
- Подготовим данные по первой гипотезе. Не будем учитывать игры с пустым рейтингом.

###### Проверим гипотезу что средние пользовательские рейтинги платформ Xbox One и PC одинаковые c альфа = 5%. Сформируем сначала нулевую гипотезу для равенства двух генеральных совокупностей и если она не опровергнется, дальше попробуем опровергнуть альтернативную гипотезу, что средние пользовательские рейтинги платформ Xbox One и PC разные.

- Нулевая гипотеза: Средние пользовательские рейтинги жанров Action и Sports одинаковые

- Альтернативна гипотеза: Средние пользовательские рейтинги жанров Action и Sports разные.

```python
stata_xboxone= period_data.query('platform == "XOne" and user_score != "NaN"')['user_score'].reset_index(drop=True).copy()
stata_pc = period_data.query('platform == "PC" and user_score != "NaN"')['user_score'].reset_index(drop=True).copy()

mean_xboxone = stata_xboxone.mean()
mean_pc = stata_pc.mean()
```

Посмотрим и сравним дисперсию двух выборок, если она отличаеться более чем на 10% тогда в Т-тесте зададим параметр, equal_var = False(по умолчанию True) и дисперсия будет оценена для каждый выборки по отдельности, а не общей.

```python
print(stata_xboxone.var())
print(stata_pc.var())
```

Мы видим что дисперсии выборок не сильно отличаются, берем аргумент equal_var=True в Т-тесте.

```python
alpha = .05 

results = st.ttest_ind(
    stata_xboxone, 
    stata_pc, equal_var=True)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
    if mean_xboxone != mean_pc:
        print('Средние пользовательские рейтинги платформ Xbox One и PC могут отличатся',mean_xboxone,'!=',mean_pc)
```

- У нас не получилось опровергнуть нулевую гипотезу о равенстве генеральных совокупностей с альфа= 5%, соответственно по альтернативной гипотезе средние пользовательские рейтинги могут отличатся, опровергнуть ее у нас не получилось. Полученное значение p-value говорит о том, что хотя средний пользовательский рейтинг и неодинаков, с вероятностью 13% такое или большее различие можно получить случайно. Это явно слишком большая вероятность, чтобы делать вывод о значимом различии между средними пользовательскими рейтингами, но мы можем допустить такой вариант, так как он не опровергнут.

- У нас получилось не опровергнуть нулевую гипотезу о равенстве генеральных совокупностей с альфа= 5%, соответственно делать какие либо выводы по данным от разных платформ мы не можем.

- Проверим гипотезу что средние пользовательские рейтинги жанров Action и Sports разные c альфа = 5%.

```python
stata_Action= period_data.query('genre == "Action" and user_score != "NaN"')['user_score'].reset_index(drop=True).copy()
stata_Sports = period_data.query('genre == "Sports" and user_score != "NaN"')['user_score'].reset_index(drop=True).copy()

mean_Action = stata_Action.mean()
mean_Sports = stata_Sports.mean()
```

- Посмотрим и сравним дисперсию двух выборок, если она отличаеться более чем на 10% тогда в Т-тесте зададим параметр, equal_var = False(по умолчанию True) и дисперсия будет оценена для каждый выборки по отдельности, а не общей.

```python
print(stata_Action.var())
print(stata_Sports.var())
```

- Мы видим что диспеерсии выборок отличается более чем на 10%, берем аргумент equal_var=False в Т-тесте.

```python
alpha = .05 

results = st.ttest_ind(
    stata_Action, 
    stata_Sports, equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
    if mean_Action != mean_Sports:
        print('Средние пользовательские рейтинги платформ Xbox One и PC могут отличатся',mean_Action,'!=',mean_Sports)
```

- У нас не получилось опровергнуть нулевую гипотезу о равнстве генеральных совокупностей с альфа= 5%, соответственно по альтернативной гипотезе средние пользовательские рейтинги могут отличатся, опровергнуть ее у нас не получилось.
